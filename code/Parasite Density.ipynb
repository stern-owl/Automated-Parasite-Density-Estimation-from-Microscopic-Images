```python
from ultralytics import YOLO
import os, csv

# Load both models
parasite_model = YOLO("/content/drive/MyDrive/best_parasite_model.pt")
wbc_model      = YOLO("/content/drive/MyDrive/WBC_dataset/runs/WBC/train1/weights/best.pt")

# Folder to evaluate (could be all images or a test split)
image_folder = "/content/drive/MyDrive/dataset/images/val"
output_csv   = "/content/drive/MyDrive/parasite_per_wbc2.csv"
```

```python
with open(output_csv, "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerow(["Image", "Parasite_Count", "WBC_Count", "Parasites_per_WBC", "Parasite_Density_per_uL"])

    for img_name in os.listdir(image_folder):
        if not img_name.lower().endswith((".jpg",".png")):
            continue
        img_path = os.path.join(image_folder, img_name)

        # Predict
        p_res = parasite_model.predict(source=img_path, save=False, conf=0.25)[0]
        w_res = wbc_model.predict(source=img_path, save=False, conf=0.25)[0]

        p_count = len(p_res.boxes)
        w_count = len(w_res.boxes)

        # Compute ratios
        parasites_per_wbc = p_count / w_count if w_count else None
        parasite_density = parasites_per_wbc * 8000 if parasites_per_wbc is not None else None

        writer.writerow([img_name, p_count, w_count, parasites_per_wbc, parasite_density])
        if parasite_density is not None:
          print(f"{img_name}: {p_count} parasites / {w_count} WBCs → Density = {parasite_density:.2f} per µL")
        else:
          print(f"{img_name}: {p_count} parasites / {w_count} WBCs → Density = N/A (no WBCs detected)")

        # print(f"{img_name}: {p_count} parasites / {w_count} WBCs → Density = {parasite_density:.2f} per µL")
```

```python
import cv2
from ultralytics import YOLO
import matplotlib.pyplot as plt

def visualize_detections(image_path, parasite_model, wbc_model, save_path=None):
    # Load image
    image = cv2.imread(image_path)
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Run inference
    p_result = parasite_model(image_path, conf=0.25)[0]
    w_result = wbc_model(image_path, conf=0.25)[0]

    # Draw parasite boxes in RED
    for box in p_result.boxes:
        x1, y1, x2, y2 = map(int, box.xyxy[0])
        cv2.rectangle(image_rgb, (x1, y1), (x2, y2), (255, 0, 0), 2)
        cv2.putText(image_rgb, "Parasite", (x1, y1 - 5), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 1)

    # Draw WBC boxes in GREEN
    for box in w_result.boxes:
        x1, y1, x2, y2 = map(int, box.xyxy[0])
        cv2.rectangle(image_rgb, (x1, y1), (x2, y2), (0, 255, 0), 2)
        cv2.putText(image_rgb, "WBC", (x1, y1 - 5), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 1)

    # Display
    plt.figure(figsize=(10, 10))
    plt.imshow(image_rgb)
    plt.axis("off")
    plt.title(f"Parasites: {len(p_result.boxes)}, WBCs: {len(w_result.boxes)}\nEstimated Density: {(len(p_result.boxes)/len(w_result.boxes))*8000:.2f}/µL" if len(w_result.boxes) else "WBCs not detected")
    plt.show()

    # Optionally save the image
    if save_path:
        cv2.imwrite(save_path, cv2.cvtColor(image_rgb, cv2.COLOR_RGB2BGR))
```

```python
from ultralytics import YOLO

parasite_model = YOLO("/content/drive/MyDrive/best_parasite_model.pt")
wbc_model = YOLO("/path/to/wbc_model.pt")

visualize_detections("path/to/image.jpg", parasite_model, wbc_model)
```
